```python
#Q7.
num = int(input("Enter a number: "))
count = 0
if num == 0:
    count = 1
else:
    while num != 0:
        num //= 10
        count += 1
print("Number of digits:", count)
```

```python
#Q8.
total = 0
num = int(input("Enter a number: "))
while num != 0:
    total += num
    num = int(input("Enter a number: "))
print("Total =", total)
print("stop")
```

```python
#Q9.
num = 101
while True:
    if num % 7 == 0:
        print("First multiple of 7 greater than 100 is:", num)
        break
    num += 1
```

```python
#Q10.
secret = 25
guess = 0
attempts = 0
while guess != secret:
    guess = int(input("Guess the number: "))
    attempts += 1
    if guess > secret:
        print("Too high")
    elif guess < secret:
        print("Too low")
print("Correct! Attempts:", attempts)
```
