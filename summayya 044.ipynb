```python
#Q6
def factorial (n):
  if n==0 or n==1:
    return 1
  return n*factorial (n-1)
```

```python
print(factorial(6))
```

```python
#Q7
def apply_function (func,nums):
  return [func(x)for x in nums]
def sq(x):
  return x**2
```

```python
print(apply_function(sq,[2,9,8,7]))
```

```python
#Q9
def apply_operation(func,nums):
  return[func(n)for n in nums]
def add_two(x):
  return x+2
```
